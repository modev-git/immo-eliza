Step 1 : Data Cleaning
You have collected your data! So it's time to do a cleaning on it. A cleaned dataset is a dataset that doesn't contain any duplicates, is blank spaces or error-free. The rest of the analysis can be discarded if you neglect this step!

No duplicates
No blank spaces (ex: " I love python " => "I love python")
No errors
No empty values

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/processed/listings.csv', skipinitialspace=True, na_values=['None'])
```

```python
#drop duplicates
df = df.drop_duplicates()
# check the number. should be zero
df.duplicated().sum()
```

69% swimming pool misding value. Not relevant = drop column? fill with 0?
77% garden area missing value. fill with 0 because apartements dont have gardens
61% terrace area missing. fill with 0
67% fully equipped kitchen. Missing value doesnt mean it doesnt have a kitchen, just not provided. fill with zero ?

```python
#fill swiming pool values with 0 for missing values
df['swimming_pool'] = df['swimming_pool'].fillna(0)

#fill garden area missing values with zero.
df['garden'] = df['garden'].fillna(0)
#df['garden_area_m2'] = df['garden_area_m2'].fillna(0) #what if it has a garden but not its surface?
# only fill garden_area_m2 with 0 if there is no garden
df.loc[df['garden'] == 0, 'garden_area_m2'] = df.loc[df['garden'] == 0, 'garden_area_m2'].fillna(0)


#fill terrace area missing values with zero.
df['terrace'] = df['terrace'].fillna(0)
#df['terrace_area_m2'] = df['terrace_area_m2'].fillna(0) #some have terrace but not terrace area
df.loc[df['terrace'] == 0, 'terrace_area_m2'] = df.loc[df['terrace'] == 0, 'terrace_area_m2'].fillna(0)
```
