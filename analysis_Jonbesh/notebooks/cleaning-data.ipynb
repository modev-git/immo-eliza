```python
import pandas as pd 
import numpy as np
```

```python
df= pd.read_csv('/Users/jonbeshahmadzai/Downloads/immovlan_dataset.csv')
df = df.drop(columns=["Open fire", "Surface area of the plot of land"])
```

Dropping duplicates:

```python
df= df.drop_duplicates()
```

Removing empty spaces: 

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

Checking Prices for error and urealistic prices:

```python
#replacing unrealistic prices with None (the limit i use is 1000 euros after checking the data and putting the price in ascending order)
df.loc[df['Price'] <=1000, 'Price'] = np.nan
```

Checking Number of Rooms (Error) and Unrealistic Number of Rooms : Wohoooooo

```python
df[df["Number of rooms"] == 0].shape
```

```python
df[df["Number of rooms"] == 0].head(10)
```

```python
df[df["Number of rooms"] == 0]["Subtype of property"].value_counts()
```

```python
#now cleaning number of rooms with 0 for Apartment, Residence, mixed building, master house, triplex ,groundfloor and penthouse. If it’s really a studio, the subtype should say Studio. When a penthouse could have 0 roomsThat would mean a penthouse studio. 

df.loc[(df["Number of rooms"] == 0) & (df["Subtype of property"] != "Studio"), "Number of rooms"] = np.nan
```

Now checking fully equiped kitchen. so since its a binary value between 1 = equiped and 0 no equiped. we can fill the missing values if its not mentioned in the listing as 0 = no equiped 

```python
df['Fully equipped kitchen'] = df['Fully equipped kitchen'].fillna(0)
```

Now we do the same thing for furnished: assuming if they didn't mention it , its not equiped

```python
df['Furnished'] = df['Furnished'].fillna(0)
```

Now Checking Living Area for Errors and Unrealistic data

```python
df.sort_values("Living Area", ascending=False)[["Subtype of property","Living Area","Price"]].head(10)
```

```python
#cleaning unrealistic vlaues for living area putting a maximum and then a minimum 
# Replacing unrealistic values (<10 m² or >5000 m²) with NaN
# These values likely cameme from scraping errors or misreported data

df.loc[df['Living Area'] > 10000 , 'Living Area'] = np.nan

df.loc[df['Living Area'] < 10 , 'Living Area'] = np.nan

```

```python
df["Living Area"].describe()
```

Now we do the same for Terrace : assuming if they didn't mention it , there is no terrace : filling the missing values with 0 as its binary. 

```python
df['Terrace'] = df['Terrace'].fillna(0)
```

Now checking Terrace Area for Errors and Unrealistic Data and also if there is no terrace, the area cannot exist.

so Setting the terrace area to missing when there is no terrace.

```python
df.loc[df["Terrace"] == 0, "Terrace Area"] = np.nan
```

```python
df.sort_values("Terrace Area", ascending=False)[["Locality","Subtype of property","Terrace Area","Price"]].head(10)
```

Also filling the Garden for missing Values: if there is a missing value then we fill it up with 0 assuming no garden :

```python
df['Garden']= df['Garden'].fillna(0)
```

Now Checking Garden Area For Errors and Unrealistic Data 

also gotta do the same for if garden = 0 then garden area = None 

```python
df.loc[df["Garden"] == 0, "Garden Area"] = np.nan
```

```python
df.sort_values("Garden Area", ascending=False)[["Locality","Subtype of property","Garden Area","Price"]].head(10)
```

```python
df.sort_values("Garden Area", ascending=True)[["Locality","Subtype of property","Garden Area","Price"]].head(10)
```

```python
# Cleaning Garden Area
# Two clearly unrealistic values were detected during inspection:
# 1 m² (impossible garden) or maybe ???? gotta check
# 690330 m² (69 hectares, likely scraping error)
# These values are replaced with NaN

df.loc[df["Garden Area"] < 2, "Garden Area"] = np.nan
df.loc[df["Garden Area"] > 500000, "Garden Area"] = np.nan
```

```python
df["Garden Area"].describe()
```

Now dealing with Total Surface of the land if it has urealistic and errored data 

```python
df.sort_values("Surface of the land", ascending=False)[["Locality","Subtype of property","Surface of the land","Price"]].head(10)
```

```python
df.sort_values("Surface of the land", ascending=True)[["Locality","Subtype of property","Surface of the land","Price"]].head(50)
```

```python
# Cleaning Surface of the land
# Values smaller than 10 m² are unrealistic for land area
# These are replaced with NaN

df.loc[df["Surface of the land"] < 10, "Surface of the land"] = np.nan
```

Now we have to Deal with Number of Facades and its Errors and unrealistic data 

```python
df.sort_values('Number of facades', ascending= False)[['Locality', 'Number of facades', 'Subtype of property', 'Price']].head()
```

```python
# Cleaning unrealistic number of facades
# In Belgian real estate properties have between 1 and 4 facades.
# Values above 4 are likely scraping errors.

df.loc[df["Number of facades"] > 4, "Number of facades"] = np.nan
```

```python
df['Number of facades'].describe()
```

Now we come to the swimming pool part: Here i will try to fill the value of None to 0 indicating there is no swimming pool if they didnt mention it :  In real estate datasets, missing usually means there is no pool because agents almost always advertise pools.


```python
df['Swimming pool']= df['Swimming pool'].fillna(0)
```
